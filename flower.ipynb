```python
import sklearn
from sklearn.datasets import load_iris
import pandas as pd
```

```python
dataset = load_iris() #step 1 : load the dataset
```

```python
# step 2:creating dataframe now
data = pd.DataFrame(dataset.data,columns = dataset.feature_names)
data['target'] = dataset.target

print(data.head())
```

```python
#checking for missing values
print(data.isnull().sum())
```

```python
# cheching the shape and type
print(data.shape)
print(type(data))
```

```python
#checking unique value in target column
print(data['target'].value_counts())
```

```python
#basic statistics
print(data.describe())
```

```python
#step3: spliting the dataset sparated features(X) and target(y)
X = data.drop('target',axis = 1)
y = data['target']

from sklearn.model_selection import train_test_split
X_train ,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state =24)
```

```python
#step 4: train the model (fitting)
from sklearn.linear_model import LogisticRegression

#creat model
model = LogisticRegression(max_iter = 1000)

#treain the model
model.fit(X_train,y_train)
```

```python
print("iterations used",model.n_iter_)
```

```python
from mmap import ALLOCATIONGRANULARITY
#step 4: make predictions
y_pred = model.predict(X_test)
#convert numerical labels to species names

prediction_species = dataset.target_names[y_pred[:10]]
acual_species = dataset.target_names[y_test[:10]]
print("predicted labels :", prediction_species)
print("actual labels:",acual_species)
```

```python
print(len(y_test))  #number of test samples used
```

```python
#step 5: a.Accuracy  on the test set

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
print("accuracy score:",accuracy)
```

```python
#b.cross validation Accuracy
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,X,y,cv =5)
print("cross validation scores:",scores)
```

```python
# step 6: predicting output on custom input
import numpy as np
#custom flower measurements 4 features
new_data = np.array([[5.1,3.5,1.4,0.2]])

#predict the class(species)
prediction = model.predict(new_data)
prediction = dataset.target_names[prediction]

print("predected class:",prediction)
```
